```python
bins = [0, 5, 13, 17, 26, 40, float('inf')]
labels = ['Baby', 'Kid', 'Teenager', 'Young Adult', 'Adult', 'Senior']

df['AgeCategory'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Add 'Unknown' for missing values
df['AgeCategory'] = df['AgeCategory'].cat.add_categories(['Unknown'])
df['AgeCategory'] = df['AgeCategory'].fillna('Unknown')
df['AgeCategory'].value_counts()
```

```python
import re

def get_title(name):
    if pd.isna(name):
        return 'Neither'
    
    # Case-insensitive and precise matching
    if re.search(r'\bMrs\b', name, flags=re.IGNORECASE):
        return 'Mrs'
    elif re.search(r'\bMiss\b', name, flags=re.IGNORECASE):
        return 'Miss'
    elif re.search(r'\bMr\b', name, flags=re.IGNORECASE):
        return 'Mr'
    elif re.search(r'\bMaster\b', name, flags=re.IGNORECASE):
        return 'Master'
    elif re.search(r'\bDr\b', name, flags=re.IGNORECASE):
        return 'Dr'
    else:
        return 'Neither'

df['Title'] = df['Name'].apply(get_title)

for t in ['Mr', 'Mrs', 'Miss', 'Master', 'Dr', 'Neither']:
    subset = df[df['Title'] == t]
    print(f"\n=== Title: {t} â€” total rows: {len(subset)} ===")
    print("\nAgeCategory counts:")
    print(subset['AgeCategory'].value_counts(dropna=False).to_string())
```

```python
# ...existing code...
# new cell - min/max age per title (and Neither)
titles = ['Mr', 'Mrs', 'Miss', 'Master', 'Dr']
masks = {t: df['Name'].str.contains(r'\b' + t + r'\b', na=False) for t in titles}
masks['Neither'] = ~df['Name'].str.contains(r'\b(?:Mr|Mrs|Miss|Master)\b', na=False)

stats = {}
for k, mask in masks.items():
    ages = df.loc[mask, 'Age'].dropna()
    stats[k] = {'count': int(mask.sum()), 'min_age': float(ages.min()) if not ages.empty else None, 'max_age': float(ages.max()) if not ages.empty else None}

stats_df = pd.DataFrame.from_dict(stats, orient='index')[['count', 'min_age', 'max_age']]
print(stats_df)
# ...existing code...
```

```python
# ...existing code...
# new cell - count and mean age per title (and Neither)
titles = ['Mr', 'Mrs', 'Miss', 'Master', 'Dr']
masks = {t: df['Name'].str.contains(r'\b' + t + r'\b', na=False) for t in titles}
masks['Neither'] = ~df['Name'].str.contains(r'\b(?:Mr|Mrs|Miss)\b', na=False)

stats = {}
for k, mask in masks.items():
    ages = df.loc[mask, 'Age'].dropna()
    stats[k] = {'count': int(mask.sum()), 'mean_age': float(ages.mean()) if not ages.empty else None}

pd.DataFrame.from_dict(stats, orient='index')[['count', 'mean_age']]
# ...existing code...
```

```python
# First, filter only rows with non-null Age
filtered_df = df[df['Age'].notnull()]

# Then, count survivors (Survived == 1) in each AgeCategory
survivor_counts = filtered_df[filtered_df['Survived'] == 1]['AgeCategory'].value_counts()

# Compare survivors vs non-survivors per category (warning-free)
survival_summary = (
    filtered_df.groupby('AgeCategory', observed=True)['Survived']
    .value_counts()
    .unstack(fill_value=0)
)

print("=== Survivors by AgeCategory ===")
print(survivor_counts.to_string())

print("\n=== Full survival summary (0 = died, 1 = survived) ===")
print(survival_summary)
```
